```python
import yfinance as yf
import matplotlib.pyplot as plt
# Получение данных для акций Apple
data = yf.download('AAPL', '2020-01-01', '2025-02-25')
# Визуализация скорректированной цены закрытия
data['Close'].plot()
plt.title('Скорректированная цена закрытия Microsoft')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.show()
```
