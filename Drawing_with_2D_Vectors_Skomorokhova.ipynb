# **Рисование с помощью двухмерных векторов**

**Задача 1: Построение точек динозавра:** создайте Python-программу для визуализации динозавра, используя точки и векторы.

```python
from matplotlib import pyplot as plt
# Список векторов, определяющих динозавра
dino_vectors = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4),

                (-5, 3), (-5, 2), (-2, 2), (-5, 1), (-4, 0), (-2, 1),

                (-1, 0), (0, -3), (-1, -4), (1, -4), (2, -3), (1, -2),

                (3, -1), (5, 1)]
# Разделяем координаты x и y
x_coords, y_coords = zip(*dino_vectors)
# Построение графика
plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, color="green", label="Точки динозавра")
plt.title("Точки контура динозавра")
plt.xlabel("X-координаты")
plt.ylabel("Y-координаты")
plt.grid(True)
plt.legend()
plt.show()
# Создаем фигуру для графика с заданными размерами
# Строим рассеянный график (scatter plot) с координатами динозавра
# Устанавливаем заголовок графика
# Устанавливаем метку для оси X
# Устанавливаем метку для оси Y
# Включаем сетку на графике для лучшей читаемости
# Добавляем легенду для графика
# Отображаем график
```

**Задача 2: Построение линий для соединения точек динозавра.** Добавьте линии для соединения точек, чтобы визуализировать контур динозавра.

```python
# Добавляем соединяющие линии
plt.figure(figsize=(8, 6))
plt.plot(*zip(*dino_vectors), linestyle="-", color="orange", label="Контур динозавра")
plt.scatter(x_coords, y_coords, color="red", label="Точки динозавра")
plt.title("Контур динозавра с точками")
plt.xlabel("X-координаты")
plt.ylabel("Y-координаты")
plt.grid(True)
plt.legend()
plt.show()
# Создаем фигуру для графика с заданными размерами
# Строим линию, соединяющую точки динозавра, с заданным стилем и цветом
# Строим рассеянный график (scatter plot) с координатами динозавра, выделяя точки красным цветом
# Устанавливаем заголовок графика
# Устанавливаем метку для оси X
# Устанавливаем метку для оси Y
# Включаем сетку на графике для лучшей читаемости
# Добавляем легенду для графика, чтобы обозначить линии и точки
# Отображаем график
```

**Вращение динозавра**

```python
from math import radians, cos, sin
# Функция вращения вектора на заданный угол theta
def rotate(vector, theta):
    angle = radians(theta) # Преобразуем угол из градусов в радианы
    return (vector[0] * cos(angle) - vector[1] * sin(angle), # Новая X-координата после вращения
            vector[0] * sin(angle) + vector[1] * cos(angle)) # Новая Y-координата после вращения

# Вращение фигуры
theta = 45
rotated_dino = [rotate(v, theta) for v in dino_vectors] # Применяем функцию вращения ко всем вектором динозавра
# Визуализация
plt.figure(figsize=(8, 6)) # Создаем фигуру для графика с заданными размерами
plt.plot(*zip(*dino_vectors), linestyle="-", color="blue", label="Исходный динозавр") # Строим линию для исходного динозавра с заданным стилем и цветом
plt.plot(*zip(*rotated_dino), linestyle="--", color="teal", label=f"Динозавр, вращённый на {theta}°") # Строим линию для вращённого динозавра с заданным стилем и цветом
plt.title("Вращение динозавра") # Устанавливаем заголовок графика
plt.legend() # Добавляем легенду для графика, чтобы обозначить линии
plt.grid(True) # Включаем сетку на графике для лучшей читаемости
plt.show() # Отображаем график
```

**Задача 3: Построение графика функции**

```python
import numpy as np # Импортируем библиотеку NumPy для работы с массивами и математическими функциями
x = np.arange(-10, 11, 1) # Создаем массив x с значениями от -10 до 10 с шагом 1
y = x**2 # Вычисляем значения y как квадрат значений x
plt.figure(figsize=(8, 6)) # Создаем фигуру для графика с заданными размерами
plt.plot(x, y, marker="o", color="teal", label="y = x^2") # Строим график функции y = x^2 с маркерами и заданным цветом
plt.title("График функции y = x^2") # Устанавливаем заголовок графика
plt.xlabel("X") # Устанавливаем метку для оси X
plt.ylabel("Y") # Устанавливаем метку для оси Y
plt.axhline(0, color="black",linewidth=0.5) # Добавляем горизонтальную линию на уровне 0 по оси Y
plt.axvline(0, color="black",linewidth=0.5) # Добавляем вертикальную линию на уровне 0 по оси X
plt.grid(True) # Включаем сетку на графике для лучшей читаемости
plt.legend() # Добавляем легенду для графика, чтобы обозначить линию
plt.show() # Отображаем график
```

**Задача 4: Перемещение динозавра**

```python
# Функция для сдвига векторов на заданные значения dx и dy
def shift_vectors(vectors, dx, dy):
    return [(x + dx, y + dy) for x, y in vectors] # Возвращаем новый список векторов, сдвинутых на dx и dy
shifted_dino = shift_vectors(dino_vectors, -2, -2) # Сдвигаем векторы динозавра на -2 по обеим осям
# Отображение исходного и сдвинутого динозавров
plt.figure(figsize=(8, 6))
plt.plot(*zip(*dino_vectors), linestyle="-", color="blue", label="Исходный динозавр") # Строим линию для исходного динозавра с заданным стилем и цветом
plt.plot(*zip(*shifted_dino), linestyle="--", color="red", label="Сдвинутый динозавр") # Строим линию для сдвинутого динозавра с заданным стилем и цветом
plt.scatter(*zip(*dino_vectors), color="blue") # Добавляем маркеры для исходного динозавра
plt.scatter(*zip(*shifted_dino), color="red") # Добавляем маркеры для сдвинутого динозавра
plt.title("Перемещение динозавра") # Устанавливаем заголовок графика
plt.xlabel("X") # Устанавливаем метку для оси X
plt.ylabel("Y") # Устанавливаем метку для оси Y
plt.grid(True) # Включаем сетку на графике для лучшей читаемости
plt.legend() # Добавляем легенду для графика, чтобы обозначить линии
plt.show()
```

**Задача 5: Построение сотни динозавров**

```python
def create_dino_grid(vectors, grid_size=5, spacing=12):

    dinos = []

    for i in range(-grid_size, grid_size):

        for j in range(-grid_size, grid_size):

            dinos.append(shift_vectors(vectors, i * spacing, j * spacing))

    return dinos


grid_dinos = create_dino_grid(dino_vectors)


# Построение всех динозавров

plt.figure(figsize=(10, 10))

for dino in grid_dinos:

    plt.plot(*zip(*dino), linestyle="-", color="teal")


plt.title("Сетка из динозавров")

plt.xlabel("X")

plt.ylabel("Y")

plt.grid(False)

plt.show()
```

**Задача 6: сложение векторов**

```python

# Функция сложения векторов

def add(v1, v2):

    return (v1[0] + v2[0], v1[1] + v2[1])


# Исходные векторы

v1 = (4, 3)

v2 = (-1, 1)

v_sum = add(v1, v2)


# Построение векторов

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, *v1, angles="xy", scale_units="xy", scale=1, color="purple", label="Вектор v1 (4, 3)")

plt.quiver(0, 0, *v2, angles="xy", scale_units="xy", scale=1, color="teal", label="Вектор v2 (-1, 1)")

plt.quiver(0, 0, *v_sum, angles="xy", scale_units="xy", scale=1, color="plum", label="Сумма v1 + v2 (3, 4)")


# Настройка осей

plt.xlim(-2, 5)

plt.ylim(-2, 5)

plt.axhline(0, color="black",linewidth=0.5)

plt.axvline(0, color="black",linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Сложение векторов")

plt.show()
```

**Задача 7: Умножение вектора на скаляр**

```python

# Функция скалярного умножения

def scale(scalar, v):

    return (scalar * v[0], scalar * v[1])


# Исходный вектор

v = (2, 3)

scaled_v = scale(2.5, v)


# Построение векторов

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1, color="red", label="Вектор v (2, 3)")

plt.quiver(0, 0, *scaled_v, angles="xy", scale_units="xy", scale=1, color="blue", label="Скалярное произведение 2.5v (5, 7.5)")


# Настройка осей

plt.xlim(-1, 6)

plt.ylim(-1, 8)

plt.axhline(0, color="black",linewidth=0.5)

plt.axvline(0, color="black",linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Умножение вектора на скаляр")

plt.show()
```

**Задача 8: Построение длины вектора**

```python
from math import sqrt


# Функция для вычисления длины вектора

def length(v):

    return sqrt(v[0]**2 + v[1]**2)


# Исходный вектор

v = (4, 3)

v_len = length(v)


print(f"Длина вектора {v}: {v_len:.2f}")
```

**Задача 9: Смещение точки**

```python
# Функция для смещения точки

def translate(translation, points):

    return [(p[0] + translation[0], p[1] + translation[1]) for p in points]


# Исходные точки

points = [(1, 2), (3, 4), (5, 6)]

translation = (-2, -1)

shifted_points = translate(translation, points)


# Вывод результата

print(f"Исходные точки: {points}")

print(f"Смещённые точки: {shifted_points}")
```

**Задача 10: Построение параллелограмма**

```python
# Функция построения параллелограмма

def parallelogram(v1, v2):

    return [v1, add(v1, v2), v2, (0, 0), v1]


# Исходные векторы

u = (3, 2)

v = (1, 4)

parallelogram_points = parallelogram(u, v)


# Построение

plt.figure(figsize=(6, 6))

plt.plot(*zip(*parallelogram_points), color="pink", linestyle="-", marker="o", label="Параллелограмм")

plt.quiver(0, 0, *u, angles="xy", scale_units="xy", scale=1, color="blue", label="Вектор u")

plt.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1, color="green", label="Вектор v")


# Настройка осей

plt.xlim(-1, 6)

plt.ylim(-1, 6)

plt.axhline(0, color="black",linewidth=0.5)

plt.axvline(0, color="black",linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Параллелограмм по двум векторам")

plt.show()
```

**Задача 11: Вычитание векторов**

```python
# Функция для вычитания векторов

def subtract(v1, v2):

    return (v1[0] - v2[0], v1[1] - v2[1])


# Исходные векторы

v1 = (5, 4)

v2 = (2, 3)

v_diff = subtract(v1, v2)


# Визуализация

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, *v1, angles="xy", scale_units="xy", scale=1, color="pink", label="Вектор v1 (5, 4)")

plt.quiver(0, 0, *v2, angles="xy", scale_units="xy", scale=1, color="red", label="Вектор v2 (2, 3)")

plt.quiver(0, 0, *v_diff, angles="xy", scale_units="xy", scale=1, color="teal", label="Разность v1 - v2 (3, 1)")


# Настройка осей

plt.xlim(-1, 6)

plt.ylim(-1, 6)

plt.axhline(0, color="black", linewidth=0.5)

plt.axvline(0, color="black", linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Вычитание векторов")

plt.show()
```

**Задача 12: Расстояние между двумя точками**

```python
# Функция для вычисления расстояния

def distance(p1, p2):

    return sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)


# Исходные точки

p1 = (1, 2)

p2 = (4, 6)

dist = distance(p1, p2)


print(f"Расстояние между точками {p1} и {p2}: {dist:.2f}")
```

**Задача 13: Длина стрелки**

```python
# Исходный вектор

v = (7, 4)

v_len = length(v)


# Построение графика

plt.figure(figsize=(6, 6))

plt.quiver(0, 0, *v, angles="xy", scale_units="xy", scale=1, color="teal", label=f"Вектор v ({v[0]}, {v[1]})")

plt.text(v[0] / 2, v[1] / 2, f"Длина = {v_len:.2f}", fontsize=10, color="teal")


# Настройка осей

plt.xlim(0, 8)

plt.ylim(0, 5)

plt.axhline(0, color="black", linewidth=0.5)

plt.axvline(0, color="black", linewidth=0.5)

plt.grid(True)

plt.gca().set_aspect('equal', adjustable='box')

plt.legend()

plt.title("Длина вектора")

plt.show()
```
